```python
import numpy as np
import pandas as pd
import plotly.express as px
import msicpe.ssl as msi

from GeneSinus import GeneSin

A = 10
phi0 = 45

t ,s = GeneSin(A , 20 , phi0 , 1.25 , 1000 , 12.5)

# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':s})
# II) Creer sa figure
fig = px.scatter(df, x='temps', y='signal')
# III) Creer son layout
fig.update_layout(xaxis_title='temps', yaxis_title='signal',
#xaxis_range=..., yaxis_range=..., 
title = 'Tracé du signal',
template='plotly_white', width=500, height=300)
# IV) Afficher
fig.show()

tf ,nu = msi.TransFourier( s, t) 

# I) Creer sa structure de données
df = pd.DataFrame({'temps':nu, 'signal':np.abs(tf)})
# II) Creer sa figure
fig = px.line(df, x='temps', y='signal')
# III) Creer son layout
fig.update_layout(xaxis_title='fréquence', yaxis_title='tf du signal',
#xaxis_range=..., yaxis_range=..., 
title = 'Tracé du tf du signal',
template='plotly_white', width=500, height=300)
# IV) Afficher
fig.show()
```
