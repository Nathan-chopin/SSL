```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as alg

# Taille de la matrice
m, n = 20, 10

# Construction de la matrice binaire
A = np.random.randint(0, 2, (m, n))

# Affichage de l'image correspondant à la matrice A (0->noir, 1->blanc)
plt.figure(1,figsize=(12, 4))
plt.subplot(131)
plt.imshow(A, cmap='gray')
plt.title("a. Image d'origine (A)", fontsize=14)

# Décomposition en valeurs singulières de A (utiliser numpy.linalg.svd)
U, S, VT = alg.svd(A)

# Affichage des valeurs singulières
plt.subplot(132)
plt.plot(S, '*k-')
plt.grid(True)
plt.title("b. Valeurs singulières", fontsize=14)

# Nombre de valeurs singulières non nulles
r = alg.matrix_rank(A)

# Reconstruction de la matrice avec les valeurs singulières principales
B = np.zeros_like(A, dtype=float)   # initialisation
for k in range(len(S) - 3):
    B += S[k] * U[:,k:k+1] @ VT[k : k+1,:]

# Affichage de l'image correspondant à la matrice B
plt.subplot(133)
plt.imshow(B, cmap='gray')
plt.title("c. Image reconstruite (B)", fontsize=14)
plt.show() 
```
