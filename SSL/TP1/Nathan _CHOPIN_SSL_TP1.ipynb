```python
# fichier python du TP1 de SSL
# Nathan CHOPIN
# 7/11/24

import numpy as np
import pandas as pd
import plotly.express as px

# fonction porte de T
def porte(t,T):
    y=(abs(t)<T/2)
    return np.float16(y)

D = 5 # s
N = 2000
t = np.linspace(-D,D,N,endpoint=False) # = 1/N * np.arange(N) -D
T = 2

# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':porte(t,T)})

# II) Creer sa figure
fig = px.line(df, x='temps', y='signal')

# III) Creer son layout
fig.update_layout(xaxis_title='Argument 1', yaxis_title='Argument 2',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé de la porte',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()
```

```python
import msicpe.ssl as msi

tf, nu = msi.TransFourier(porte(t,T),t)

# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':porte(t,T), 'frequence' : nu, 'tf_real' : tf.real, 'tf_imag' : tf.imag})

# II) Creer sa figure
fig2 = px.line(df, x='frequence', y='tf_imag')
fig1 = px.line(df, x='frequence', y='tf_real')

# III) Creer son layout
fig1.update_layout(yaxis_title='transformer de fourier réelle', xaxis_title='fréquence (Hz)',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé de partie réelle',
template='plotly_white', width=500, height=300)

fig2.update_layout(yaxis_title='transformer de fourier imaginaire', xaxis_title='fréquence (Hz)',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé de partie imaginaire',
template='plotly_white', width=500, height=300)


# IV) Afficher
fig1.show()
fig2.show()
```

```python

t0 = 2 #s

porte_decaler = porte(t - t0,T)

#######################superposition des portes
# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':porte(t,T), 'signal_decaler' : porte_decaler, 'frequence' : nu, 'tf_real' : tf.real, 'tf_imag' : tf.imag})

# II) Creer sa figure
fig = px.line(df, x='temps', y=['signal','signal_decaler'])

# III) Creer son layout
fig.update_layout(yaxis_title='Argument 1', xaxis_title='temps(s)',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé des portes',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()

tf_decaler, nu = msi.TransFourier(porte_decaler,t)

#######################superposition tf
# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':porte(t,T), 'frequence' : nu, 'tf' : tf , 'tf_real' : tf.real, 'tf_imag' : tf.imag,
'signal_decaler' : porte_decaler, 'tf_decaler' : tf_decaler, 'tf_real_decaler' : tf_decaler.real, 'tf_imag_decaler' : tf_decaler.imag})

# II) Creer sa figure
fig = px.line(df, x='frequence', y=['tf_real','tf_real_decaler'])

# III) Creer son layout
fig.update_layout(yaxis_title='Argument 1', xaxis_title='fréquence (Hz)',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé des partie réelle',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()

#######################superposition partie imag
# I) Creer sa structure de données
df = pd.DataFrame({'temps':t, 'signal':porte(t,T), 'frequence' : nu, 'tf' : tf , 'tf_real' : tf.real, 'tf_imag' : tf.imag,
'signal_decaler' : porte_decaler, 'tf_decaler' : tf_decaler, 'tf_real_decaler' : tf_decaler.real, 'tf_imag_decaler' : tf_decaler.imag})

# II) Creer sa figure
fig = px.line(df, x='frequence', y=['tf_imag','tf_imag_decaler'])

# III) Creer son layout
fig.update_layout(yaxis_title='Argument 1', xaxis_title='fréquence (Hz)',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Tracé des partie imag',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()
```
