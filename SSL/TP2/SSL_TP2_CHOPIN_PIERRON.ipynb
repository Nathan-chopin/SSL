```python
# fichier python du TP2 de SSL
# Nathan CHOPIN et Mal PIERRON
# 19/11/24

import numpy as np
import pandas as pd
import plotly.express as px

# fonction porte de T
def porte(t,T):
    y=(abs(t)<T/2)
    return np.float16(y)

A = 1
Fs = 1000 #Hz
Ts = 1/Fs # s
f0 = 7 #Hz
T0 = 1 / f0
##################################################
T = 5 * T0
D = 3 * T # s
t = np.arange(0,D,Ts)
N = len(t)
##################################################

tau0 = 1 # s

x = A *np.cos(2 * np.pi * f0 * t) * porte(t - T/2 , T)

y = A *np.cos(2 * np.pi * f0 * t - tau0) * porte(t -tau0 - T/2 , T)

# I) Creer sa structure de donnes
df = pd.DataFrame({'temps':t, 'signal':x , 'signal retour' :y})

# II) Creer sa figure
fig = px.line(df, x='temps', y=['signal','signal retour'])

# III) Creer son layout
fig.update_layout(xaxis_title='temps(s)', yaxis_title='signaux',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Trac des signaux',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()
```

```python
gamma_xy = np.correlate(y,x,mode='full')
tau = []
for k in range(-(N-1),N):
    tau.append(k*Ts)

# I) Creer sa structure de donnes
df = pd.DataFrame({'temps':tau, 'signal':gamma_xy })

# II) Creer sa figure
fig = px.line(df, x='temps', y='signal')

# III) Creer son layout
fig.update_layout(xaxis_title='tau', yaxis_title='gamma_xy',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Trac de l\'intercorrelation',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()

indice_tau_min = np.argmax(- gamma_xy)
indice_tau_max = np.argmax(gamma_xy)

tau_0moy = (tau[indice_tau_min] + tau[indice_tau_max] )/ 2
print('tau exp = ',tau_0moy ,'s')
print('% erreur :',abs(tau0 - tau_0moy) * 100,'%')
```

```python
sigma = 10

z = y + sigma * np.random.randn(np.size(y))

# I) Creer sa structure de donnes
df = pd.DataFrame({'temps':t, 'signal':x , 'signal retour' :z})

# II) Creer sa figure
fig = px.line(df, x='temps', y=['signal','signal retour'])

# III) Creer son layout
fig.update_layout(xaxis_title='temps(s)', yaxis_title='signaux',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Trac des signaux',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()



#intercorrelation :

gamma_xz = np.correlate(z,x,mode='full')

# I) Creer sa structure de donnes
df = pd.DataFrame({'temps':tau, 'gammaxz':gamma_xz ,'gammaxy':gamma_xy})

# II) Creer sa figure
fig = px.line(df, x='temps', y=['gammaxz','gammaxy'])

# III) Creer son layout
fig.update_layout(xaxis_title='tau', yaxis_title='gamma',
#xaxis_range=..., yaxis_range=..., # if needed
title = 'Trac de l\'intercorrelation',
template='plotly_white', width=500, height=300)

# IV) Afficher
fig.show()
```
