En calculant Ã  la main le minimum, on trouve x* = (-1, 3/2) et donc f(x*) = -5/4.

```python
import numpy as np
import numpy.linalg as alg
import pandas as pd
import plotly.express as px

def fonction(x, y):
    return x - y + 2 * x**2 + 2 * x * y + y**2

def gradient(x, y):
    df_dx = 4 * x + 2 * y + 1
    df_dy = 2 * x + 2 * y - 1
    return np.array([df_dx, df_dy])

def descente_gradient(point_depart, pas, precision):
    point_actuel = np.array(point_depart)
    chemin = [point_actuel]
    while np.linalg.norm(gradient(point_actuel[0], point_actuel[1])) >= precision:
        grad = gradient(point_actuel[0], point_actuel[1])
        point_actuel = point_actuel - pas * grad
        chemin.append(point_actuel)
    return np.array(chemin)

depart = [3, 4]
pas = 0.1
precision = 1e-2

chemin = descente_gradient(depart, pas, precision)
minimum = chemin[-1]
minth = np.array([-1,1.5])
delta = alg.norm(minth - minimum)/alg.norm(minth)
print(f"l'erreur relative est delta = {delta}.")
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = fonction(X, Y)

surface_donnees = pd.DataFrame({
    'x': X.ravel(),
    'y': Y.ravel(),
    'z': Z.ravel()
})

chemin_donnees = pd.DataFrame({
    'x': chemin[:, 0],
    'y': chemin[:, 1],
    'z': [fonction(x, y) for x, y in chemin]
})

fig = px.scatter_3d(
    surface_donnees,
    x='x',
    y='y',
    z='z',
    opacity=0.6,
    color='z',
    color_continuous_scale='Viridis',
    title="Surface et chemin de descente"
)

fig.add_scatter3d(
    x=chemin_donnees['x'],
    y=chemin_donnees['y'],
    z=chemin_donnees['z'],
    mode='markers+lines',
    marker=dict(size=5, color='red'),
    line=dict(color='orange', width=2),
    name="Chemin de descente"
)

fig.add_scatter3d(
    x=[minimum[0]],
    y=[minimum[1]],
    z=[fonction(minimum[0], minimum[1])],
    mode='markers',
    marker=dict(size=8, color='green'),
    name='Minimum trouvÃ©'
)

fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=800,
    height=800
)
```
