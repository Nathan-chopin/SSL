```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, a, b, c):
    return a * x**2 + b * x + c

def df(x, a, b): 
    return 2 * a * x + b

xmin, xmax = -2, 3
x = np.linspace(xmin, xmax, 1000)

a, b, c = 2, -3, 1
y = f(x, a, b, c)

x_th = -b / (2 * a)

x0 = 3
lambda_ = 0.1
epsilon = 1e-3
def descente_grad(x0,lambda_,epsilon):
    xk = x0
    iterations = 0

    while df(xk, a, b) > epsilon:
        xk = xk - lambda_ * df(xk, a, b)
        iterations += 1
    delta = abs((x_th - xk)/x_th)
    return delta, iterations, xk

delta, iterations, x_min_grad = descente_grad(x0, lambda_, epsilon)
plt.figure(figsize=(10, 6))
plt.plot(x, y,color='blue')
plt.scatter([x_th], [f(x_th, a, b, c)], color='red')
plt.scatter([x_min_grad], [f(x_min_grad, a, b, c)], color='green')
print(delta)
```

```python
Delta, Iterations, Lambda = [], [], []
for lambdav in range(5, 400):
    lambdav /= 100 
    Lambda.append(lambdav)
    delta, iterations, xk = descente_grad(x0, lambdav, epsilon)
    Delta.append(delta)
    Iterations.append(iterations)

plt.figure(figsize=(12, 6))

plt.plot(Lambda, Delta, color="blue")
plt.plot(Lambda, Iterations, color="green")
```
